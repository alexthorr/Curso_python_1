###  What is a Dictionary in Python?

A **dictionary** in Python is a **collection of keyvalue pairs**, where:

- Each key must be **unique** and **immutable** (like a string, number, or tuple).  
- Each value can be of **any type** (number, string, list, even another dictionary).  
- Dictionaries are **mutable**, so you can **add, remove, or modify** elements.  
- Dictionaries are **unordered** in older versions of Python (before 3.7).  
  From Python 3.7+, they **maintain insertion order**.


```python
# Basic syntax
my_dict = {
    "key1": "value1",
    "key2": "value2"
}
```
